```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a figure and 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Define a line
P0 = np.array([1, 2, 3])  # A point on the line
V = np.array([2, 1, -1])  # Direction vector of the line

# Parametric equation of the line
t = np.linspace(0, 3, 100)
line = P0 + np.outer(t, V)

# Plot the line
ax.plot3D(line[:, 0], line[:, 1], line[:, 2], label='Line')

# Define a plane
Q0 = np.array([0, 0, 0])  # A point on the plane
N = np.array([1, -1, 2])  # Normal vector to the plane

# Create a meshgrid for the plane
x, y = np.meshgrid(np.linspace(-5, 5, 20), np.linspace(-5, 5, 20))
z = (-N[0] * (x - Q0[0]) - N[1] * (y - Q0[1]) + Q0[2]) / N[2]

# Plot the plane
ax.plot_surface(x, y, z, alpha=0.5, color='cyan', label='Plane')

# Set labels
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

# Add a legend
ax.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the two line equations
def line1(x):
    return 2 * x + 1

def line2(x):
    return -1.5 * x + 3

# Generate x values
x_values = np.linspace(-5, 5, 100)

# Calculate y values for each line
y_values_line1 = line1(x_values)
y_values_line2 = line2(x_values)

# Plot the lines
plt.plot(x_values, y_values_line1, label='2x + 1')
plt.plot(x_values, y_values_line2, label='-1.5x + 3')

# Set labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Plotting Two Lines on the X-axis')

# Add a legend
plt.legend()

# Show the plot
plt.grid(True)
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Create a figure and 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Define two lines in 3D space
def line1(t):
    return t, 2*t + 1, 0

def line2(t):
    return t, -1.5*t + 3, 0

# Parameter values for the lines
t_values = np.linspace(-5, 5, 100)

# Plot the lines in 3D space
ax.plot3D(*line1(t_values), label='Line 1')
ax.plot3D(*line2(t_values), label='Line 2')

# Project the lines onto the x-axis (2D projection)
ax.plot(line1(t_values)[0], np.zeros_like(t_values), np.zeros_like(t_values), linestyle='--', color='gray')
ax.plot(line2(t_values)[0], np.zeros_like(t_values), np.zeros_like(t_values), linestyle='--', color='gray')

# Set labels
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

# Add a legend
ax.legend()

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Data points for Place 1
place1_data = np.array([[0, 0, 0], [0, 1, 10], [0, 2, 12], [0, 3, 11]])

# Data points for Place 2
place2_data = np.array([[1, 0, 0], [1, 1, 8], [1, 2, 10], [1, 3, 9]])

# Extracting x, y, and z coordinates
place1_x, place1_y, place1_z = place1_data[:, 0], place1_data[:, 1], place1_data[:, 2]
place2_x, place2_y, place2_z = place2_data[:, 0], place2_data[:, 1], place2_data[:, 2]

# Fit a line through the data points for Place 1
place1_line = np.polyfit(place1_x, place1_z, 1)
place1_line_equation = np.poly1d(place1_line)

# Fit a line through the data points for Place 2
place2_line = np.polyfit(place2_x, place2_z, 1)
place2_line_equation = np.poly1d(place2_line)

# Create a figure and 3D axis
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the data points for Place 1
ax.scatter(place1_x, place1_y, place1_z, label='Place 1 Data', marker='o')

# Plot the best-fit line for Place 1
ax.plot(place1_x, place1_y, place1_line_equation(place1_x), label='Place 1 Line', color='r')

# Plot the data points for Place 2
ax.scatter(place2_x, place2_y, place2_z, label='Place 2 Data', marker='o')

# Plot the best-fit line for Place 2
ax.plot(place2_x, place2_y, place2_line_equation(place2_x), label='Place 2 Line', color='b')

# Set labels
ax.set_xlabel('X-axis (Places)')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

# Add a legend
ax.legend()

# Show the plot
plt.show()
```
