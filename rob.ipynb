```python
import pandas as pd

import yfinance as yf

amzn = yf.Ticker("AMZN")

# get all stock info
amzn.info

# get historical market data
hist = amzn.history(period="1mo")

# show meta information about the history (requires history() to be called first)
amzn.history_metadata

# show actions (dividends, splits, capital gains)
amzn.actions
amzn.dividends
amzn.splits
amzn.capital_gains  # only for mutual funds & etfs

# show share count
amzn.get_shares_full(start="2022-01-01", end=None)

# show financials:
# - income statement
amzn.income_stmt
amzn.quarterly_income_stmt
# - balance sheet
amzn.balance_sheet
amzn.quarterly_balance_sheet
# - cash flow statement
amzn.cashflow
amzn.quarterly_cashflow
# see `Ticker.get_income_stmt()` for more options

# show holders
amzn.major_holders
amzn.institutional_holders
amzn.mutualfund_holders
amzn.insider_transactions
amzn.insider_purchases
amzn.insider_roster_holders

# show recommendations
amzn.recommendations
amzn.recommendations_summary
amzn.upgrades_downgrades

# Show future and historic earnings dates, returns at most next 4 quarters and last 8 quarters by default. 
# Note: If more are needed use amzn.get_earnings_dates(limit=XX) with increased limit argument.
amzn.earnings_dates

# show ISIN code - *experimental*
# ISIN = International Securities Identification Number
amzn.isin

# show options expirations
amzn.options

# show news
amzn.news



```
